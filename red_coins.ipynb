```python
from skimage import io, filters
%matplotlib inline
```

```python
def get_avg_red(data):
    red = 0
    for r in range(0, data.shape[0]):
        for c in range(0, data.shape[1]):
            red += data[r][c][0]
    red = red / (data.shape[0] * data.shape[1])
    return(red)
```

```python
imname = 'images/IMG_0013.JPG'
```

```python
test_foto = io.imread(imname)
```
